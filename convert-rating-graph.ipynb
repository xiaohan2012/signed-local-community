```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_csv('graphs/bitcoin/soc-sign-bitcoinotc.csv', header=None, names=['u', 'v', 'rating', 'time'])
```

```python
df['sign'] = (df['rating'] > 0).apply(lambda f: 1 if f else -1)
```

```python
verbose = True
g = nx.Graph()
for _, (u, v, sign) in df[['u', 'v', 'sign']].iterrows():
    if g.has_edge(u, v):
        if g[u][v]['sign'] != sign:
            if verbose > 0:
                print('conflicting sign for edge ({}, {}), remove it'.format(u, v))
            g.remove_edge(u, v)
    g.add_edge(u, v)
    g[u][v]['sign'] = sign
g.remove_edges_from(nx.selfloop_edges(g))
g = nx.convert_node_labels_to_integers(g)
```

```python
nx.write_gpickle(g, 'graphs/bitcoin.pkl')
```
